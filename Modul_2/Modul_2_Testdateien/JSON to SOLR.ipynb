```python
url = "https://www.bibsonomy.org/json/search/Christentum?items=1000&duplicates=merged"
```

```python
import requests
result = requests.get(url) 
data = result.json()
```

```python
edited_items = [] # Auswahl der Informationen, die ausgewählt werden sollen:
desired_entries = ['type', 'year', 'publisher', 'authors', 'label', 'url', 'tags', 'doi', 'abstract']
```

```python
# für jedes Element im eingelesenen JSON file... (über data.keys() lässt sich leicht herausfinden, wo die relevanten Daten stecken):

for item in data["items"]:     # erstelle ein neues leeres dictionary für die Daten des Elements
    if item['type'] == 'Publication':
        item_export = {}  # für jeden gewünschten Eintrag
    # hier können Filterkriterien hinzugefügt werden (hier z.B. erschienen nach 2009 und mit weniger als 4 Autoren):
   # if (item["year"])>2009  and len(item["author"])<4: 
           
        for entry in desired_entries:  # ... überprüfe, ob es einen Eintrag gibt (um Fehler zu vermeiden)

            if entry in item.keys():   # füge den Eintrag den Daten zum Element hinzu

                item_export[entry] = item[entry] 
                
        edited_items.append(item_export)    # am Ende: füge den vollständigen (gefilterten) Eintrag der zu exportierenden Liste hinzu
```

```python
for item in edited_items:
    if 'author' in item.keys():
        for el in item.get('author'):
            print(el)
```

```python
for item in edited_items:
    if 'tags' in item.keys():
        for el in item.get('tags'):
            print(el)
```
