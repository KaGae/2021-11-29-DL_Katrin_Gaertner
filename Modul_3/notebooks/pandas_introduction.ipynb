```python
df['Total Checkouts'].mean() # Mittelwert = Durchschnitt ist in der Statistik das arithmetische Mittel. 
#Errechnet hat ihn schon jeder: Man addiert die Werte, deren Mittelwert gesucht wird 
#und teilt sie durch ihre Anzahl.
```

```python
df['Total Checkouts'].quantile(q=[0.25, 0.5, 0.75]) # Quantile = definiert einen bestimmten Teil einer Datenmenge,
#das heißt, ein Quantil legt fest, wie viele Werte einer Verteilung über oder unter einer bestimmten Grenze liegen. 
#Besondere Quantile sind das Quartil (Viertel), das Quintil (Fünftel) und das Percentil (Hundertstel).
```

```python
# Ausreißerentfernung
# identifies 0.5% of the data at both ends of the distribution 
alpha = 0.005
df['Total Checkouts'].quantile([alpha, 1-alpha])
```

#### 3.5 Exkurs: Ausreißerentfernung II (30 Min)
###### Identifiziere positive Ausreißer in der Spalte Total Checkouts. Ausreißer werden jetzt als Beobachtungen xi definiert, für die gilt:
###### xi>x0.75+1.5xIQR
###### Das heißt, eine Beobachtungen gilt als Ausreißer, wenn sie größer als die Summe aus dem 75% Quantil und dem 1.5-fachen des Interquartilsabstands ist.
###### Wie viel Prozent der Beobachtungen im Datensatz werden mit dieser Methode als Ausreißer markiert?

```python
IQR = df['Total Checkouts'].quantile(q=0.75) - df['Total Checkouts'].quantile(q=0.25)
IQR
```

```python
# Berechnung des Quantiles 
qt_75 = df['Total Checkouts'].quantile(q=0.75)
qt_75
```

```python
outlier_frontier_2 = (qt_75 + 1.5 * IQR)
outlier_frontier_2
```

```python
df_outlier_2 = df[df['Total Checkouts']< outlier_frontier_2]
df_outlier_2
```

```python
df['Total Checkouts'].count()
```
